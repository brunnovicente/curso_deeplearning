#Obtenção dos Dados

```
import pandas as pd
import numpy as np

dados = pd.read_csv('drive/My Drive/basesinhas/casas.csv')
dados
```

#Limpeza dos Dados

```
dados = dados[['Lot Area','Gr Liv Area','Garage Area','Overall Cond','Year Built','Yr Sold', 'SalePrice']]
dados
```

```
dados.isna().sum()
```

```
dados = dados.dropna()
```

```
X = dados.drop(['SalePrice'], axis=1)
y = dados['SalePrice']

from sklearn.preprocessing import MinMaxScaler
sca = MinMaxScaler()

X = sca.fit_transform(X)
X[0:5,:]
```

```
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7)
```

#Treino

```
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor(n_neighbors=10)
knn.fit(X_train, y_train)
```

#Teste

```
y_pred = knn.predict(X_test)
```

#Avaliação

```
from sklearn.metrics import mean_absolute_error

erro = mean_absolute_error(y_test, y_pred)
erro
```
