```
import pandas as pd
import numpy as np
```

```
dados = pd.read_csv('drive/My Drive/basesinhas/xor.csv')
X = dados.drop(['y'], axis=1).values
y = dados['y'].values
X
```

```
class Perceptron:

    def __init__(self):
        self.w = np.array([-5., 3.5])
        self.n = 0.01 #taxa de aprendizagem
    
    def ativacao(self, x):
        if x < 1:
            return 0
        else:
            return 1
    
    def somatorio(self, x):
        return np.dot(self.w, x)
    
    def predizer(self, x):
        return self.ativacao(self.somatorio(x))

    def treinar(self, X, y):

        while True:
            erroTotal = 0
            print(self.w)
            for i, x in enumerate(X):
                y_pred = self.predizer(x)
                erro = y[i] - y_pred

                for j, wj in enumerate(self.w):
                    self.w[j] = wj + self.n * erro * x[j]
                erroTotal += np.abs(erro)

            if erroTotal == 0:
                break

```

```
neuronio = Perceptron()
neuronio.treinar(X, y)


for xi in X:
    print(neuronio.predizer(xi))
```
